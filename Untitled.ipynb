```python
import os
import glob
import json
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import scipy.sparse as sparse
import networkx as nx  # Added to fix your visualization error

import torch
import torch.nn.functional as F
from torch.utils.data import DataLoader, Dataset
from torch_geometric.data import HeteroData
from torch_geometric.nn import HeteroConv, GCNConv, Linear

# Metrics
from sklearn.model_selection import train_test_split
from sklearn.metrics import roc_curve, auc

device = torch.device('cuda' if torch.cuda.is_available() else 'cpu')
print(f"Using device: {device}")
```

```python
FOLDER_PATH = "/home/juan/Work/Midterm project/splited"  # Keep your original path
X_PATH = os.path.join(FOLDER_PATH, "X_data.npz")
Y_PATH = os.path.join(FOLDER_PATH, "Y_data.npz")
COLS_PATH = os.path.join(FOLDER_PATH, "feature_names.json")

# Only process if files don't exist
if not os.path.exists(X_PATH) or not os.path.exists(Y_PATH):
    print("Processing CSVs into Sparse Matrices...")
    FILE_PATTERN = os.path.join(FOLDER_PATH, "processed_*.csv")
    files = sorted(glob.glob(FILE_PATTERN))

    # Use first file to determine column structure
    f_df = pd.read_csv(files[0], nrows=3)
    cols = f_df.columns
    x_cols = [c for c in cols if c.startswith("x ")]
    y_cols = [c for c in cols if c.startswith("y ")]

    with open(COLS_PATH, 'w') as f:
        json.dump(x_cols, f)

    X_list = []
    Y_list = []

    for f in files:
        df = pd.read_csv(f)
        x_data = df[x_cols].fillna(0).values.astype(np.float32)
        y_data = df[y_cols].fillna(0).values.astype(np.float32)
        X_list.append(sparse.csr_matrix(x_data))
        Y_list.append(sparse.csr_matrix(y_data))

    X_final = sparse.vstack(X_list)
    Y_final = sparse.vstack(Y_list)

    sparse.save_npz(X_PATH, X_final)
    sparse.save_npz(Y_PATH, Y_final)
    print("Processing Complete.")
else:
    print("Sparse files found. Skipping processing.")
```

```python
# Load Data
X = sparse.load_npz(X_PATH)
Y = sparse.load_npz(Y_PATH)

num_patients = X.shape[0]
num_input_features = X.shape[1]  # Number of drugs
num_reactions = Y.shape[1]       # Number of reactions

print(f"Patients: {num_patients}")
print(f"Drugs (Features): {num_input_features}")
print(f"Reactions (Labels): {num_reactions}")

# Train/Test Split
indices = np.arange(num_patients)
train_idx, test_idx = train_test_split(indices, test_size=0.2, random_state=1)
```

```python
class HeteroGraphDataset(Dataset):
    def __init__(self, X, Y, indices):
        self.X = X
        self.Y = Y
        self.indices = indices

    def __len__(self):
        return len(self.indices)

    def __getitem__(self, idx):
        return self.indices[idx]

def graph_collate_fn(batch_indices):
    batch_indices = sorted(batch_indices)
    
    # Extract sub-matrices for this batch
    x_sub = X[batch_indices]
    y_sub = Y[batch_indices] 
    
    batch_size = len(batch_indices)

    # 1. Build Patient-Drug Edges
    rows_d, cols_d = x_sub.nonzero()
    edge_index_drug = torch.stack([torch.from_numpy(rows_d), torch.from_numpy(cols_d)]).long()
    
    # 2. Build Patient-Reaction Edges
    rows_r, cols_r = y_sub.nonzero()
    edge_index_react = torch.stack([torch.from_numpy(rows_r), torch.from_numpy(cols_r)]).long()
    
    data = HeteroData()
    
    # Nodes
    data['patient'].x = torch.ones((batch_size, 1), dtype=torch.float) # Dummy feature
    data['patient'].num_nodes = batch_size
    data['drug'].num_nodes = num_input_features
    data['reaction'].num_nodes = num_reactions
    
    # Edges
    data['patient', 'takes', 'drug'].edge_index = edge_index_drug
    data['patient', 'has_reaction', 'reaction'].edge_index = edge_index_react
    
    # Reverse Edges (Crucial for information flow back to patient)
    data['drug', 'taken_by', 'patient'].edge_index = torch.flip(edge_index_drug, [0])
    data['reaction', 'reaction_in', 'patient'].edge_index = torch.flip(edge_index_react, [0])
    
    return data

# Data Loaders
batch_size = 1024 
train_ds = HeteroGraphDataset(X, Y, train_idx)
train_loader = DataLoader(train_ds, batch_size=batch_size, shuffle=True, collate_fn=graph_collate_fn, num_workers=0)

test_ds = HeteroGraphDataset(X, Y, test_idx)
test_loader = DataLoader(test_ds, batch_size=batch_size, shuffle=True, collate_fn=graph_collate_fn, num_workers=0)
```

```python
from torch_geometric.nn import HeteroConv, GraphConv, Linear

class HeteroGNN(torch.nn.Module):
    def __init__(self, hidden_channels, out_channels, metadata):
        super().__init__()

        # Layer 1: GraphConv supports bipartite graphs (Patient -> Drug)
        self.conv1 = HeteroConv({
            edge_type: GraphConv(hidden_channels, hidden_channels, aggr='mean')
            for edge_type in metadata[1]
        }, aggr='sum') 

        # Layer 2: Output layer
        self.conv2 = HeteroConv({
            edge_type: GraphConv(hidden_channels, out_channels, aggr='mean')
            for edge_type in metadata[1]
        }, aggr='sum')

    def forward(self, x_dict, edge_index_dict):
        x = self.conv1(x_dict, edge_index_dict)
        x = {k: v.relu() for k, v in x.items()}
        x = self.conv2(x, edge_index_dict)
        return x

# Schema Definition (Same as before)
metadata = (
    ['patient', 'drug', 'reaction'],
    [('patient', 'takes', 'drug'), 
     ('drug', 'taken_by', 'patient'),
     ('patient', 'has_reaction', 'reaction'),
     ('reaction', 'reaction_in', 'patient')]
)

# Initialize Model
model = HeteroGNN(hidden_channels=64, out_channels=64, metadata=metadata).to(device)

# Initialize Embeddings (Same as before)
drug_emb = torch.nn.Embedding(num_input_features + 1, 64).to(device)
react_emb = torch.nn.Embedding(num_reactions + 1, 64).to(device)
patient_lin = Linear(1, 64).to(device)

# Optimizer
optimizer = torch.optim.Adam(
    list(model.parameters()) + list(drug_emb.parameters()) + list(react_emb.parameters()) + list(patient_lin.parameters()), 
    lr=0.01
)

print("HeteroGNN (GraphConv) Model Initialized successfully.")
```

```python
THRESHOLD = 0.8
eps = 1e-5
history = {'ts' : [], 'tf1' : []}
prev_avg_loss = float('inf')
```

```python
print("Starting Training...")
while True:
    model.train()
    total_loss = 0
    total_examples = 0
    total_tp, total_fp, total_fn = 0, 0, 0

    for i, batch in enumerate(train_loader):
        batch = batch.to(device)
        optimizer.zero_grad()

        # Prepare Input Features (Embeddings)
        x_dict = {
            'patient' : patient_lin(batch['patient'].x),
            'drug' : drug_emb(torch.arange(num_input_features, device=device)), 
            'reaction' : react_emb(torch.arange(num_reactions, device=device))
        }

        # Forward Pass
        out = model(x_dict, batch.edge_index_dict)

        # Link Prediction (Contrastive Learning)
        # 1. Get embedding of patients and their ACTUAL reactions
        edge_index = batch['patient', 'has_reaction', 'reaction'].edge_index
        if edge_index.numel() == 0: continue

        src_emb = out['patient'][edge_index[0]]
        dst_emb = out['reaction'][edge_index[1]]
        pos_score = (src_emb * dst_emb).sum(dim=-1) # Dot product similarity

        # 2. Get embedding of patients and RANDOM reactions (Negative Sampling)
        neg_dst_idx = torch.randint(0, num_reactions, (len(pos_score),), device=device)
        neg_dst_emb = out['reaction'][neg_dst_idx]
        neg_score = (src_emb * neg_dst_emb).sum(dim=-1)

        # 3. Calculate Loss
        scores = torch.cat([pos_score, neg_score])
        labels = torch.cat([torch.ones_like(pos_score), torch.zeros_like(neg_score)])
        
        loss = F.binary_cross_entropy_with_logits(scores, labels)

        loss.backward()
        optimizer.step()

        # Metrics Tracking
        with torch.no_grad():
            pos_probs = pos_score.sigmoid()
            neg_probs = neg_score.sigmoid()

            tp = (pos_probs > THRESHOLD).sum().item()
            fn = (pos_probs <= THRESHOLD).sum().item()
            fp = (neg_probs > THRESHOLD).sum().item()

            total_tp += tp
            total_fn += fn
            total_fp += fp
            
            total_loss += loss.item() * batch['patient'].num_nodes
            total_examples += batch['patient'].num_nodes

    # Epoch Summary
    avg_loss = total_loss / total_examples
    loss_diff = abs(prev_avg_loss - avg_loss)
    
    epsilon = 1e-9
    precision = total_tp / (total_tp + total_fp + epsilon)
    recall = total_tp / (total_tp + total_fn + epsilon)
    f1 = 2 * (precision * recall) / (precision + recall + epsilon)

    history['ts'].append(avg_loss)
    history['tf1'].append(f1)

    print(f"loss {avg_loss:.4f} | Prec: {precision:.4f} | Rec: {recall:.4f} | F1: {f1:.4f}")

    if loss_diff < eps:
        print(f"Converged. Final Loss: {avg_loss}")
        break
    
    prev_avg_loss = avg_loss
```

```python
model.eval()
total_test_tp = 0
total_test_fp = 0
total_test_fn = 0

all_y_true = []
all_y_scores = []

print("Evaluating on Test Set...")
with torch.no_grad():
    for batch in test_loader:
        batch = batch.to(device)
        
        x_dict = {
            'patient': patient_lin(batch['patient'].x),
            'drug': drug_emb(torch.arange(num_input_features, device=device)),
            'reaction': react_emb(torch.arange(num_reactions, device=device))
        }
        
        out = model(x_dict, batch.edge_index_dict)
        
        edge_index = batch['patient', 'has_reaction', 'reaction'].edge_index
        if edge_index.numel() == 0: continue

        src_emb = out['patient'][edge_index[0]]
        dst_emb = out['reaction'][edge_index[1]]
        
        pos_score = (src_emb * dst_emb).sum(dim=-1)
        
        neg_dst_idx = torch.randint(0, num_reactions, (len(pos_score),), device=device)
        neg_dst_emb = out['reaction'][neg_dst_idx]
        neg_score = (src_emb * neg_dst_emb).sum(dim=-1)
        
        pos_probs = pos_score.sigmoid()
        neg_probs = neg_score.sigmoid()
        
        # Store for ROC Curve
        batch_scores = torch.cat([pos_probs, neg_probs]).cpu().numpy()
        batch_labels = torch.cat([torch.ones_like(pos_probs), torch.zeros_like(neg_probs)]).cpu().numpy()
        all_y_scores.append(batch_scores)
        all_y_true.append(batch_labels)

        # Calculate confusion matrix parts
        tp = (pos_probs > THRESHOLD).sum().item()
        fn = (pos_probs <= THRESHOLD).sum().item()
        fp = (neg_probs > THRESHOLD).sum().item() 
        
        total_test_tp += tp
        total_test_fp += fp
        total_test_fn += fn

all_y_true = np.concatenate(all_y_true)
all_y_scores = np.concatenate(all_y_scores)

epsilon = 1e-9
test_precision = total_test_tp / (total_test_tp + total_test_fp + epsilon)
test_recall = total_test_tp / (total_test_tp + total_test_fn + epsilon)
test_f1 = 2 * (test_precision * test_recall) / (test_precision + test_recall + epsilon)

print(f"Test Precision: {test_precision:.4f}")
print(f"Test Recall:    {test_recall:.4f}")
print(f"Test F1-Score:  {test_f1:.4f}")
```

```python
fpr, tpr, thresholds = roc_curve(all_y_true, all_y_scores)
roc_auc = auc(fpr, tpr)

plt.figure(figsize=(8, 5))
plt.plot(fpr, tpr, color='darkorange', lw=2, label=f'ROC curve (area = {roc_auc:.4f})')
plt.plot([0, 1], [0, 1], color='navy', lw=2, linestyle='--', label='Random Guess')
plt.xlim([0.0, 1.0])
plt.ylim([0.0, 1.05])
plt.xlabel('Sensivity')
plt.ylabel('1 - Specificity')
plt.title('ROC')
plt.savefig('roc_curve.png')
plt.show()
```

```python
def visualize_schema():
    G_schema = nx.DiGraph()
    
    G_schema.add_node("Patient", color='skyblue')
    G_schema.add_node("Drug", color='lightgreen')
    G_schema.add_node("Reaction", color='salmon')
    
    G_schema.add_edge("Patient", "Drug", label="takes")
    G_schema.add_edge("Drug", "Patient", label="taken by")
    G_schema.add_edge("Patient", "Reaction", label="has_reaction")
    G_schema.add_edge("Reaction", "Patient", label="reaction in")
    
    plt.figure(figsize=(6, 4))
    pos = {
        'Patient': (0.5, 0.5),
        'Drug': (0.2, 0.2),
        'Reaction': (0.8, 0.2)
    }
    
    node_colors = [nx.get_node_attributes(G_schema, 'color')[n] for n in G_schema.nodes()]
    
    nx.draw(G_schema, pos, with_labels=True, node_color=node_colors, 
            node_size=3000, font_weight='bold', arrowsize=20, arrows=True)
    
    edge_labels = nx.get_edge_attributes(G_schema, 'label')
    nx.draw_networkx_edge_labels(G_schema, pos, edge_labels=edge_labels, font_color='red')
    
    plt.axis('off')
    plt.savefig('hg1.png')
    plt.show()

visualize_schema()
```

```python
import networkx as nx
import matplotlib.pyplot as plt
import random

def visualize_example_subgraph(dataset, num_samples=3):
    """
    Visualizes the connections for a few random patients from the dataset.
    """
    # 1. Pick random patients
    sample_indices = random.sample(range(len(dataset)), num_samples)
    
    # 2. Create a NetworkX Graph
    G = nx.DiGraph()
    
    # 3. Add nodes and edges for each sampled patient
    for i, patient_idx in enumerate(sample_indices):
        patient_node_id = f"Patient {patient_idx}"
        G.add_node(patient_node_id, color='skyblue', type='patient')
        
        # Get patient's data row (Sparse Matrix row)
        # dataset.X is the Drug Matrix, dataset.Y is the Reaction Matrix
        # We need to access the raw sparse matrix row for this patient
        
        # Extract Drugs (non-zero entries in X)
        patient_drugs_row = dataset.X[patient_idx]
        _, drug_indices = patient_drugs_row.nonzero()
        
        for drug_idx in drug_indices:
            drug_node_id = f"Drug {drug_idx}"
            G.add_node(drug_node_id, color='lightgreen', type='drug')
            G.add_edge(patient_node_id, drug_node_id, label='takes')
            
        # Extract Reactions (non-zero entries in Y)
        patient_reactions_row = dataset.Y[patient_idx]
        _, reaction_indices = patient_reactions_row.nonzero()
        
        for react_idx in reaction_indices:
            react_node_id = f"Reaction {react_idx}"
            G.add_node(react_node_id, color='salmon', type='reaction')
            G.add_edge(patient_node_id, react_node_id, label='has')

    # 4. Draw the Graph
    plt.figure(figsize=(12, 8))
    pos = nx.spring_layout(G, k=0.3, iterations=50)  # k controls spacing
    
    # Get colors
    node_colors = [nx.get_node_attributes(G, 'color')[n] for n in G.nodes()]
    
    # Draw Nodes
    nx.draw_networkx_nodes(G, pos, node_color=node_colors, node_size=2000, alpha=0.9)
    
    # Draw Labels (IDs)
    nx.draw_networkx_labels(G, pos, font_size=10, font_weight='bold')
    
    # Draw Edges
    nx.draw_networkx_edges(G, pos, arrowstyle='->', arrowsize=20, edge_color='gray')
    
    # Draw Edge Labels (Optional - can get cluttered)
    # edge_labels = nx.get_edge_attributes(G, 'label')
    # nx.draw_networkx_edge_labels(G, pos, edge_labels=edge_labels, font_size=8)
    
    # Legend
    from matplotlib.lines import Line2D
    legend_elements = [
        Line2D([0], [0], marker='o', color='w', markerfacecolor='skyblue', label='Patient', markersize=15),
        Line2D([0], [0], marker='o', color='w', markerfacecolor='lightgreen', label='Drug', markersize=15),
        Line2D([0], [0], marker='o', color='w', markerfacecolor='salmon', label='Reaction', markersize=15)
    ]
    plt.legend(handles=legend_elements, loc='upper right')
    
    plt.title(f"Heterogeneous Graph Subgraph (Random {num_samples} Patients)")
    plt.axis('off')
    plt.tight_layout()
    plt.savefig('hg.png')
    plt.show()

# Run the visualization using your training dataset
visualize_example_subgraph(train_ds, num_samples=5)
```
